```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt # to draw graphs
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('/content/drive/MyDrive/AI와무역투자론/Data/default_credit_card.csv')
df
#pd.options.display.max_columns=50
#df
```

###  Feature Selection

```python
# predicators & target
features = ['LIMIT_BAL','SEX','EDUCATION','MARRIAGE','AGE','PAY_0','PAY_2','PAY_3','BILL_AMT1','BILL_AMT2','BILL_AMT3','PAY_AMT1','PAY_AMT2','PAY_AMT3']
label = ['default_YN']

#df1 = df[label + features]
#df1.info()
```

Train/Test DataSet Split

### Decision Tree 알고리즘 호출

```python
from sklearn.tree import DecisionTreeClassifier # to build a classification tree
```

```python
clf_dt = DecisionTreeClassifier(criterion='entropy',max_depth=6, min_samples_leaf=10)
```

### 학습(Learning)

```python
clf_dt.fit(df[features], df[label])
```

## 학습결과 시각화(Decsion Tree)

```python
from sklearn.tree import plot_tree # to draw a classification tree

plt.figure(figsize=(20,10))
plot_tree(clf_dt,
          filled=True,
          rounded=True,
          proportion=False,
          #class_names=["default_N","default_Y"],
          fontsize=10,
          feature_names=features)
plt.show()
```
